```python
from collections import defaultdict

def calculate_average_income(degrees, incomes):
    #           
    degree_income = defaultdict(list)

    #         
    for degree, income in zip(degrees, incomes):
        degree_income[degree].append(income)
    
    #      
    average_income_by_degree = {degree: sum(income_list) / len(income_list)
                                for degree, income_list in degree_income.items()}
    
    #          
    sorted_degrees = sorted(average_income_by_degree.keys(),reverse=True)
    sorted_avg_incomes = [average_income_by_degree[degree] for degree in sorted_degrees]
    
    return sorted_degrees, sorted_avg_incomes

#  
degrees = [3, 3, 3, 2, 2, 1, 1, 1, 1]  #  
incomes = [1000, 1500, 1200, 800, 850, 500, 600, 550, 450]  #  

#    
X_a,Y_a = calculate_average_income(X_a, Y_a)
X_a_C,Y_a_C = calculate_average_income(X_a_C, Y_a_C)
```

```python
from scipy.optimize import curve_fit
def exponential_func(x, a, b):
    return a * np.exp(b * x)
```

```python
x=np.array(x)
y=np.array(y)
popt, pcov = curve_fit(lambda t,a,b: a*np.exp(b*t),  x,  y)
plt.scatter(x, y, label='Data')
x_fit = np.linspace(min(x), max(x), 100)
y_fit = exponential_func(x_fit, *popt)
plt.plot(x_fit, y_fit, 'r-', label='Exponential Fit')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential Regression')
plt.show()
```

```python
import numpy as np

def gini(incomes):
    #  
    n = len(incomes)
    
    #   
    sorted_incomes = np.sort(incomes)
    
    #   
    mean_income = np.mean(sorted_incomes)
    
    #     
    income_diffs = np.abs(np.subtract.outer(sorted_incomes, sorted_incomes)).sum()
    
    #   
    gini_coefficient = income_diffs / (2 * n**2 * mean_income)
    
    return gini_coefficient

#  
#incomes = [500, 1000, 1500, 2000, 2500]  #   
gini_coefficient = gini(Y_C_O)
print(" :", gini_coefficient)
```
